##üî∫ **Audio Espa√±ol - Modelo gpt-3.5-turbo**

```python

import speech_recognition as sr
from google.colab import files


# Funci√≥n para generar la nota cl√≠nica utilizando OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica con m√°s de 30 a√±os de experiencia tomando notas de alta calidad, espec√≠ficas y usando vocabulario t√©cnico m√©dico experto.
Transcribe el siguiente texto de una consulta m√©dica en una nota de evoluci√≥n cl√≠nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si alg√∫n dato falta, indica como "No especificado".

Formato de salida:

### Nota de Evoluci√≥n Cl√≠nica

**Fecha de Consulta:** [Fecha]
**M√©dico Responsable:** [Nombre del M√©dico]

---

**Secci√≥n 1: Evaluaci√≥n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [N√∫mero de Identificaci√≥n]

2. **S√≠ntomas Descritos:**
   - [S√≠ntomas observados o informados]

3. **Ex√°menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible Diagn√≥stico:**
   - [Hip√≥tesis diagn√≥stica inicial]

---

**Secci√≥n 2: Tratamiento y Seguimiento**

5. **Diagn√≥stico Confirmado:**
   - [Diagn√≥stico confirmado despu√©s de evaluaci√≥n]

6. **Tratamientos Aplicados:**
   - [Descripci√≥n detallada de tratamientos administrados]

7. **Receta M√©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraci√≥n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y pr√≥ximos pasos]
"""

    respuesta = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica."},
            {"role": "user", "content": prompt}
        ]
    )

    nota_clinica = respuesta['choices'][0]['message']['content'].strip()
    print(nota_clinica)
    return nota_clinica

# Funci√≥n para transcribir el archivo de audio
def transcribir_audio_archivo():
    archivo_subido = files.upload()

    reconocedor = sr.Recognizer()

    for nombre_archivo in archivo_subido.keys():
        print(f"Procesando archivo: {nombre_archivo}")
        with sr.AudioFile(nombre_archivo) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        try:
            texto_transcrito = reconocedor.recognize_google(audio, language="es-ES")
            print("Transcripci√≥n completada:")
            print(texto_transcrito)
            generar_nota_clinica_openai(texto_transcrito)

        except sr.UnknownValueError:
            print("No se pudo entender el audio.")
        except sr.RequestError as e:
            print(f"Error en el servicio de reconocimiento: {e}")

# Llamada principal
transcribir_audio_archivo()
```

```python
import openai
import speech_recognition as sr
from google.colab import files
import csv
from datetime import datetime


# Funci√≥n para guardar la nota cl√≠nica en un archivo CSV
def guardar_nota_clinica_csv(nota_clinica):
    # Crear el nombre del archivo basado en fecha y hora
    fecha_actual = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"Nota_Clinica_{fecha_actual}.csv"

    # Dividir el texto de la nota en l√≠neas
    lineas = nota_clinica.splitlines()

    # Guardar en CSV (Campo, Valor)
    with open(nombre_archivo, mode="w", newline='', encoding="utf-8") as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(["Campo", "Valor"])  # Encabezados CSV

        for linea in lineas:
            if "**" in linea:  # Detectar campos principales
                campo, valor = linea.split(":", 1) if ":" in linea else (linea, "")
                escritor_csv.writerow([campo.replace("**", "").strip(), valor.strip()])
            elif "-" in linea:  # Subcampos con guiones
                subcampo = linea.replace("-", "").strip()
                escritor_csv.writerow(["", subcampo])  # Subcampos en una l√≠nea separada
            elif linea:  # Otras l√≠neas de texto
                escritor_csv.writerow(["", linea.strip()])

    # Descargar autom√°ticamente el archivo
    files.download(nombre_archivo)
    print(f"Archivo CSV guardado y descargado: {nombre_archivo}")

# Funci√≥n para generar la nota cl√≠nica utilizando OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica con m√°s de 30 a√±os de experiencia tomando notas de alta calidad, espec√≠ficas y usando vocabulario t√©cnico m√©dico experto.
Transcribe el siguiente texto de una consulta m√©dica en una nota de evoluci√≥n cl√≠nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si alg√∫n dato falta, indica como "No especificado".

Formato de salida:

### Nota de Evoluci√≥n Cl√≠nica

**Fecha de Consulta:** [Fecha]
**M√©dico Responsable:** [Nombre del M√©dico]

---

**Secci√≥n 1: Evaluaci√≥n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [N√∫mero de Identificaci√≥n]

2. **S√≠ntomas Descritos:**
   - [S√≠ntomas observados o informados]

3. **Ex√°menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible Diagn√≥stico:**
   - [Hip√≥tesis diagn√≥stica inicial]

---

**Secci√≥n 2: Tratamiento y Seguimiento**

5. **Diagn√≥stico Confirmado:**
   - [Diagn√≥stico confirmado despu√©s de evaluaci√≥n]

6. **Tratamientos Aplicados:**
   - [Descripci√≥n detallada de tratamientos administrados]

7. **Receta M√©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraci√≥n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y pr√≥ximos pasos]
"""

    # Generar respuesta usando OpenAI
    respuesta = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica."},
            {"role": "user", "content": prompt}
        ]
    )

    # Extraer y mostrar la nota cl√≠nica generada
    nota_clinica = respuesta['choices'][0]['message']['content'].strip()
    print("Nota cl√≠nica generada:")
    print(nota_clinica)

    # Guardar la nota cl√≠nica en CSV
    guardar_nota_clinica_csv(nota_clinica)
    return nota_clinica

# Funci√≥n para transcribir el archivo de audio con detecci√≥n autom√°tica de idioma
def transcribir_audio_archivo():
    archivo_subido = files.upload()

    reconocedor = sr.Recognizer()

    for nombre_archivo in archivo_subido.keys():
        print(f"Procesando archivo: {nombre_archivo}")
        with sr.AudioFile(nombre_archivo) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        try:
            # Transcribir con detecci√≥n autom√°tica de idioma
            texto_transcrito = reconocedor.recognize_google(audio)
            print("Transcripci√≥n completada:")
            print(texto_transcrito)
            generar_nota_clinica_openai(texto_transcrito)

        except sr.UnknownValueError:
            print("No se pudo entender el audio.")
        except sr.RequestError as e:
            print(f"Error en el servicio de reconocimiento: {e}")

# Llamada principal
transcribir_audio_archivo()
```

```python
import openai
import speech_recognition as sr
import gradio as gr
import csv
from datetime import datetime

# Configurar la API de OpenAI
openai.api_key = 'key'

# Funci√≥n para exportar la nota cl√≠nica en un archivo CSV
def guardar_nota_clinica_csv(nota_clinica):
    fecha_actual = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"Nota_Clinica_{fecha_actual}.csv"

    # Guardar en CSV limpio
    with open(nombre_archivo, mode="w", newline='', encoding="utf-8") as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(["Campo", "Valor"])

        for linea in nota_clinica.splitlines():
            if "**" in linea and ":" in linea:
                campo, valor = linea.split(":", 1)
                escritor_csv.writerow([campo.replace("**", "").strip(), valor.strip()])
            elif linea.startswith("-"):  # Subcampos
                subcampo = linea.replace("-", "").strip()
                escritor_csv.writerow(["", subcampo])

    return nombre_archivo

# Funci√≥n para generar la nota cl√≠nica con OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica con m√°s de 30 a√±os de experiencia tomando notas de alta calidad, espec√≠ficas y usando vocabulario t√©cnico m√©dico experto.
Transcribe el siguiente texto de una consulta m√©dica en una nota de evoluci√≥n cl√≠nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si alg√∫n dato falta, indica como "No especificado".

Formato de salida:

### Nota de Evoluci√≥n Cl√≠nica

**Fecha de Consulta:** [Fecha]
**M√©dico Responsable:** [Nombre del M√©dico]

---

**Secci√≥n 1: Evaluaci√≥n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [N√∫mero de Identificaci√≥n]

2. **S√≠ntomas Descritos:**
   - [S√≠ntomas observados o informados]

3. **Ex√°menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible Diagn√≥stico:**
   - [Hip√≥tesis diagn√≥stica inicial]

---

**Secci√≥n 2: Tratamiento y Seguimiento**

5. **Diagn√≥stico Confirmado:**
   - [Diagn√≥stico confirmado despu√©s de evaluaci√≥n]

6. **Tratamientos Aplicados:**
   - [Descripci√≥n detallada de tratamientos administrados]

7. **Receta M√©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraci√≥n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y pr√≥ximos pasos]
"""

    respuesta = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Eres un m√©dico asistente especializado en documentaci√≥n cl√≠nica."},
            {"role": "user", "content": prompt}
        ]
    )

    return respuesta['choices'][0]['message']['content'].strip()

# Funci√≥n para procesar el archivo de audio
def procesar_audio(archivo_audio):
    reconocedor = sr.Recognizer()

    try:
        # Leer el archivo de audio
        with sr.AudioFile(archivo_audio) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        # Transcribir el audio
        texto_transcrito = reconocedor.recognize_google(audio, language="es-ES")
        print("Transcripci√≥n completada:", texto_transcrito)

        # Generar la nota cl√≠nica
        nota_clinica = generar_nota_clinica_openai(texto_transcrito)
        print("Nota cl√≠nica generada:", nota_clinica)

        # Exportar a CSV
        nombre_csv = guardar_nota_clinica_csv(nota_clinica)

        return texto_transcrito, nota_clinica, f"Archivo CSV exportado: {nombre_csv}"

    except Exception as e:
        return f"Error: {str(e)}", "", ""

# Interfaz con Gradio
gr.Interface(
    fn=procesar_audio,
    inputs=gr.Audio(type="filepath", label="Sube tu archivo de audio"),
    outputs=[
        gr.Textbox(label="Transcripci√≥n del Audio"),
        gr.Textbox(label="Nota de Evoluci√≥n Cl√≠nica Generada"),
        gr.Textbox(label="Estado de Exportaci√≥n a CSV")
    ],
    title="ClinicNotesAI - Generador de Notas Cl√≠nicas",
    description="Sube un archivo de audio para transcribirlo y generar autom√°ticamente una nota cl√≠nica en formato estructurado."
).launch(share=True)
```
