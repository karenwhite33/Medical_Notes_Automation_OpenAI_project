```python

import speech_recognition as sr
from google.colab import files


# FunciÃ³n para generar la nota clÃ­nica utilizando OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica con mÃ¡s de 30 aÃ±os de experiencia tomando notas de alta calidad, especÃ­ficas y usando vocabulario tÃ©cnico mÃ©dico experto.
Transcribe el siguiente texto de una consulta mÃ©dica en una nota de evoluciÃ³n clÃ­nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si algÃºn dato falta, indica como "No especificado".

Formato de salida:

### Nota de EvoluciÃ³n ClÃ­nica

**Fecha de Consulta:** [Fecha]
**MÃ©dico Responsable:** [Nombre del MÃ©dico]

---

**SecciÃ³n 1: EvaluaciÃ³n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [NÃºmero de IdentificaciÃ³n]

2. **SÃ­ntomas Descritos:**
   - [SÃ­ntomas observados o informados]

3. **ExÃ¡menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible DiagnÃ³stico:**
   - [HipÃ³tesis diagnÃ³stica inicial]

---

**SecciÃ³n 2: Tratamiento y Seguimiento**

5. **DiagnÃ³stico Confirmado:**
   - [DiagnÃ³stico confirmado despuÃ©s de evaluaciÃ³n]

6. **Tratamientos Aplicados:**
   - [DescripciÃ³n detallada de tratamientos administrados]

7. **Receta MÃ©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraciÃ³n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y prÃ³ximos pasos]
"""

    respuesta = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica."},
            {"role": "user", "content": prompt}
        ]
    )

    nota_clinica = respuesta['choices'][0]['message']['content'].strip()
    print(nota_clinica)
    return nota_clinica

# FunciÃ³n para transcribir el archivo de audio
def transcribir_audio_archivo():
    archivo_subido = files.upload()

    reconocedor = sr.Recognizer()

    for nombre_archivo in archivo_subido.keys():
        print(f"Procesando archivo: {nombre_archivo}")
        with sr.AudioFile(nombre_archivo) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        try:
            texto_transcrito = reconocedor.recognize_google(audio, language="es-ES")
            print("TranscripciÃ³n completada:")
            print(texto_transcrito)
            generar_nota_clinica_openai(texto_transcrito)

        except sr.UnknownValueError:
            print("No se pudo entender el audio.")
        except sr.RequestError as e:
            print(f"Error en el servicio de reconocimiento: {e}")

# Llamada principal
transcribir_audio_archivo()
```

##ğŸ”º**Audio en inglÃ©s - Modelo gpt-4o - Guardar transcripciÃ³n en csv**

```python
import openai
import speech_recognition as sr
from google.colab import files
import csv
from datetime import datetime


# FunciÃ³n para guardar la nota clÃ­nica en un archivo CSV
def guardar_nota_clinica_csv(nota_clinica):
    # Crear el nombre del archivo basado en fecha y hora
    fecha_actual = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"Nota_Clinica_{fecha_actual}.csv"

    # Dividir el texto de la nota en lÃ­neas
    lineas = nota_clinica.splitlines()

    # Guardar en CSV (Campo, Valor)
    with open(nombre_archivo, mode="w", newline='', encoding="utf-8") as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(["Campo", "Valor"])  # Encabezados CSV

        for linea in lineas:
            if "**" in linea:  # Detectar campos principales
                campo, valor = linea.split(":", 1) if ":" in linea else (linea, "")
                escritor_csv.writerow([campo.replace("**", "").strip(), valor.strip()])
            elif "-" in linea:  # Subcampos con guiones
                subcampo = linea.replace("-", "").strip()
                escritor_csv.writerow(["", subcampo])  # Subcampos en una lÃ­nea separada
            elif linea:  # Otras lÃ­neas de texto
                escritor_csv.writerow(["", linea.strip()])

    # Descargar automÃ¡ticamente el archivo
    files.download(nombre_archivo)
    print(f"Archivo CSV guardado y descargado: {nombre_archivo}")

# FunciÃ³n para generar la nota clÃ­nica utilizando OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica con mÃ¡s de 30 aÃ±os de experiencia tomando notas de alta calidad, especÃ­ficas y usando vocabulario tÃ©cnico mÃ©dico experto.
Transcribe el siguiente texto de una consulta mÃ©dica en una nota de evoluciÃ³n clÃ­nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si algÃºn dato falta, indica como "No especificado".

Formato de salida:

### Nota de EvoluciÃ³n ClÃ­nica

**Fecha de Consulta:** [Fecha]
**MÃ©dico Responsable:** [Nombre del MÃ©dico]

---

**SecciÃ³n 1: EvaluaciÃ³n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [NÃºmero de IdentificaciÃ³n]

2. **SÃ­ntomas Descritos:**
   - [SÃ­ntomas observados o informados]

3. **ExÃ¡menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible DiagnÃ³stico:**
   - [HipÃ³tesis diagnÃ³stica inicial]

---

**SecciÃ³n 2: Tratamiento y Seguimiento**

5. **DiagnÃ³stico Confirmado:**
   - [DiagnÃ³stico confirmado despuÃ©s de evaluaciÃ³n]

6. **Tratamientos Aplicados:**
   - [DescripciÃ³n detallada de tratamientos administrados]

7. **Receta MÃ©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraciÃ³n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y prÃ³ximos pasos]
"""

    # Generar respuesta usando OpenAI
    respuesta = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica."},
            {"role": "user", "content": prompt}
        ]
    )

    # Extraer y mostrar la nota clÃ­nica generada
    nota_clinica = respuesta['choices'][0]['message']['content'].strip()
    print("Nota clÃ­nica generada:")
    print(nota_clinica)

    # Guardar la nota clÃ­nica en CSV
    guardar_nota_clinica_csv(nota_clinica)
    return nota_clinica

# FunciÃ³n para transcribir el archivo de audio con detecciÃ³n automÃ¡tica de idioma
def transcribir_audio_archivo():
    archivo_subido = files.upload()

    reconocedor = sr.Recognizer()

    for nombre_archivo in archivo_subido.keys():
        print(f"Procesando archivo: {nombre_archivo}")
        with sr.AudioFile(nombre_archivo) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        try:
            # Transcribir con detecciÃ³n automÃ¡tica de idioma
            texto_transcrito = reconocedor.recognize_google(audio)
            print("TranscripciÃ³n completada:")
            print(texto_transcrito)
            generar_nota_clinica_openai(texto_transcrito)

        except sr.UnknownValueError:
            print("No se pudo entender el audio.")
        except sr.RequestError as e:
            print(f"Error en el servicio de reconocimiento: {e}")

# Llamada principal
transcribir_audio_archivo()
```

```python
import openai
import speech_recognition as sr
import gradio as gr
import csv
from datetime import datetime

# Configurar la API de OpenAI
openai.api_key = 'key'

# FunciÃ³n para exportar la nota clÃ­nica en un archivo CSV
def guardar_nota_clinica_csv(nota_clinica):
    fecha_actual = datetime.now().strftime("%Y%m%d_%H%M%S")
    nombre_archivo = f"Nota_Clinica_{fecha_actual}.csv"

    # Guardar en CSV limpio
    with open(nombre_archivo, mode="w", newline='', encoding="utf-8") as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerow(["Campo", "Valor"])

        for linea in nota_clinica.splitlines():
            if "**" in linea and ":" in linea:
                campo, valor = linea.split(":", 1)
                escritor_csv.writerow([campo.replace("**", "").strip(), valor.strip()])
            elif linea.startswith("-"):  # Subcampos
                subcampo = linea.replace("-", "").strip()
                escritor_csv.writerow(["", subcampo])

    return nombre_archivo

# FunciÃ³n para generar la nota clÃ­nica con OpenAI
def generar_nota_clinica_openai(texto_transcrito):
    prompt = f"""
Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica con mÃ¡s de 30 aÃ±os de experiencia tomando notas de alta calidad, especÃ­ficas y usando vocabulario tÃ©cnico mÃ©dico experto.
Transcribe el siguiente texto de una consulta mÃ©dica en una nota de evoluciÃ³n clÃ­nica detallada dividida en dos etapas.

Texto transcrito: {texto_transcrito}

Si algÃºn dato falta, indica como "No especificado".

Formato de salida:

### Nota de EvoluciÃ³n ClÃ­nica

**Fecha de Consulta:** [Fecha]
**MÃ©dico Responsable:** [Nombre del MÃ©dico]

---

**SecciÃ³n 1: EvaluaciÃ³n Inicial**

1. **Datos del Paciente:**
   - Nombre: [Nombre del Paciente]
   - Edad: [Edad]
   - ID: [NÃºmero de IdentificaciÃ³n]

2. **SÃ­ntomas Descritos:**
   - [SÃ­ntomas observados o informados]

3. **ExÃ¡menes y Resultados:**
   - [Pruebas realizadas y hallazgos]

4. **Posible DiagnÃ³stico:**
   - [HipÃ³tesis diagnÃ³stica inicial]

---

**SecciÃ³n 2: Tratamiento y Seguimiento**

5. **DiagnÃ³stico Confirmado:**
   - [DiagnÃ³stico confirmado despuÃ©s de evaluaciÃ³n]

6. **Tratamientos Aplicados:**
   - [DescripciÃ³n detallada de tratamientos administrados]

7. **Receta MÃ©dica:**
   - Medicamentos: [Nombre, dosis, frecuencia y duraciÃ³n]

8. **Plan de Seguimiento:**
   - [Recomendaciones y prÃ³ximos pasos]
"""

    respuesta = openai.ChatCompletion.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": "Eres un mÃ©dico asistente especializado en documentaciÃ³n clÃ­nica."},
            {"role": "user", "content": prompt}
        ]
    )

    return respuesta['choices'][0]['message']['content'].strip()

# FunciÃ³n para procesar el archivo de audio
def procesar_audio(archivo_audio):
    reconocedor = sr.Recognizer()

    try:
        # Leer el archivo de audio
        with sr.AudioFile(archivo_audio) as fuente_audio:
            audio = reconocedor.record(fuente_audio)

        # Transcribir el audio
        texto_transcrito = reconocedor.recognize_google(audio, language="es-ES")
        print("TranscripciÃ³n completada:", texto_transcrito)

        # Generar la nota clÃ­nica
        nota_clinica = generar_nota_clinica_openai(texto_transcrito)
        print("Nota clÃ­nica generada:", nota_clinica)

        # Exportar a CSV
        nombre_csv = guardar_nota_clinica_csv(nota_clinica)

        return texto_transcrito, nota_clinica, f"Archivo CSV exportado: {nombre_csv}"

    except Exception as e:
        return f"Error: {str(e)}", "", ""

# Interfaz con Gradio
gr.Interface(
    fn=procesar_audio,
    inputs=gr.Audio(type="filepath", label="Sube tu archivo de audio"),
    outputs=[
        gr.Textbox(label="TranscripciÃ³n del Audio"),
        gr.Textbox(label="Nota de EvoluciÃ³n ClÃ­nica Generada"),
        gr.Textbox(label="Estado de ExportaciÃ³n a CSV")
    ],
    title="ClinicNotesAI - Generador de Notas ClÃ­nicas",
    description="Sube un archivo de audio para transcribirlo y generar automÃ¡ticamente una nota clÃ­nica en formato estructurado."
).launch(share=True)
```
